```python
def combinedata():
    combined_X=X_train.append(X_test)
    #combined_X['desc']=combinedX['desc'].astype(str)
    combined_X['diff_creat_laun']=abs(combined_X['created_at']-combined_X['launched_at'])
    combined_X['diff_laun_status']=abs(combined_X['launched_at']-combined_X['state_changed_at'])
    combined_X['diff_status_dead']=(combined_X['state_changed_at']-combined_X['deadline'])
    print("Shape of total data",combined_X.shape)
    return combined_X
```

```python
df=combinedX.keywords
```

```python
from nltk.corpus import stopwords
stop_words =(stopwords.words('english'))
stop_words.extend(['mr','my','the','a','may','also','zero','one','two','three','four','five','six','seven','eight','nine','ten','across','among','beside','however','yet','within'])
```

```python
df2=df.copy()
df2=df2.str.lower().str.split('-')
df3=[] # Removing stop words
for item in df2:
    df3.append(list(set(item)-set(stop_words)))
df4=[] # removing alpha numeric words
for item in df3:
    df4.append([x for x in item if not any(c.isdigit() for c in x)])
```

```python

df5=[]
for item in df4:
    df5.append([' '.join(item)])
```
